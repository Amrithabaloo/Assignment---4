```python
#Line Plot
price =[149,225,150,89,80,200,60,79,79,150,135,85,89,85,120,140,215,140,99]
minimum_nights =[1,1,3,1,10,3,45,2,2,1,5,2,4,2,90,2,2,1,3]
plt.plot (price,minimum_nights)
plt.xlabel('X-axis','price')
plt.ylabel('Y-axis','minimum_nights')
plt.title('Line Plot')
plt.show()
```

```python
#Bar Plot
neighbourhood =['Kensington','Midtown' ,'Harlem','Clinton Hill','East Harlem','Murray Hill','Bedford-Stuyvesant','Upper West Side']
neighbourhood_group =['Brooklyn','Manhattan','Manhattan','Brooklyn','Manhattan','Manhattan','Brooklyn','Manhattan']
plt.bar (neighbourhood_group,neighbourhood)
plt.xlabel('X-axis','neighbourhood_group')
plt.ylabel('Y-axis','neighbourhood')
plt.title('Bar Plot')
plt.show()
```

```python
#Scatter Plot
neighbourhood =['Kensington','Midtown' ,'Harlem','Clinton Hill','East Harlem','Murray Hill','Bedford-Stuyvesant','Upper West Side']
neighbourhood_group =['Brooklyn','Manhattan','Manhattan','Brooklyn','Manhattan','Manhattan','Brooklyn','Manhattan']
plt.scatter(neighbourhood_group,neighbourhood)
plt.xlabel('X-axis','neighbourhood_group')
plt.ylabel('Y-axis','neighbourhood')
plt.title('scatter Plot')
plt.show()
```

```python
#Line Plot
host_id=[2787,2845,4632,4869]
neighbourhood_group =['Brooklyn','Manhattan','Manhattan','Brooklyn']
sns.lineplot (x=host_id,y=neighbourhood_group)
plt.xlabel('X-axis','host_id')
plt.ylabel('Y-axis','neighbourhood_group')
plt.title('Line Plot')
plt.show()
```

```python
#bar Plot
host_id=[2787,2845,4632,4869]
neighbourhood_group =['Brooklyn','Manhattan','Manhattan','Brooklyn']
sns.barplot (x=host_id,y=neighbourhood_group)
plt.xlabel('X-axis','host_id')
plt.ylabel('Y-axis','neighbourhood_group')
plt.title('bar Plot')
plt.show()
```

```python
#scatter Plot
host_id=[2787,2845,4632,4869]
neighbourhood_group =['Brooklyn','Manhattan','Manhattan','Brooklyn']
sns.scatterplot (x=host_id,y=neighbourhood_group)
plt.xlabel('X-axis','host_id')
plt.ylabel('Y-axis','neighbourhood_group')
plt.title('scatter Plot')
plt.show()
```

```python
#hist Plot
host_id=[2787,2845,4632,4869]
neighbourhood_group =['Brooklyn','Manhattan','Manhattan','Brooklyn']
sns.histplot (x=host_id,y=neighbourhood_group)
plt.xlabel('X-axis','host_id')
plt.ylabel('Y-axis','neighbourhood_group')
plt.title('hist Plot')
plt.show()
```

```python
#heat Plot
host_id=[2787,2845,4632,4869]
neighbourhood_group =['Brooklyn','Manhattan','Manhattan','Brooklyn']
sns.heattplot (x=host_id,y=neighbourhood_group)
plt.xlabel('X-axis','host_id')
plt.ylabel('Y-axis','neighbourhood_group')
plt.title('heat Plot')
plt.show()
```
